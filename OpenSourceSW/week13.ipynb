```python
#실습 1
import matplotlib.pyplot as plt
import csv

f = open('seoul1.csv')

data = csv.reader(f)

next(data)

max=[]
min=[]

for row in data:
    if ((row[0].split('/')[0] == '3')and (row[0].split('/')[1]=='20')):
        max.append(float(row[-1]))
        min.append(float(row[3]))

plt.title("Visualize Birthday Temperature")
plt.plot(max)
plt.plot(min)
plt.show()
```

```python
#실습 2
import matplotlib.pyplot as plt
import csv

f = open('seoul1.csv')

data = csv.reader(f)

next(data)

aug=[]
jan=[]
for row in data:
    month = row[0].split('/')[0]
    if row[-1] !='':
        if month=='8':
            aug.append(float(row[-1]))
        elif month=='1':
            jan.append(float(row[-1]))

plt.hist(jan, bins = 100, color='b', label='Jan')            
plt.hist(aug, bins = 100, color='r', label='Aug')
plt.legend()
plt.show()
```
